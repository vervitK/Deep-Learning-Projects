```python
import os
import cv2
from google.colab.patches import cv2_imshow
import pandas as pd


meta_data = pd.DataFrame(columns=['x','y','w','h', 'Total_Faces','Image_Name'])
CClassifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

image_directory = '/content/Train'

image_files = os.listdir(image_directory)

for i in image_files:
    if i.endswith(('.jpg')):
        file_path = os.path.join(image_directory, i)
        image = cv2.imread(file_path)
        g_scale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        face = CClassifier.detectMultiScale(g_scale, scaleFactor=1.1, minNeighbors=5, minSize=(250, 250))
        for (x, y, w, h) in face:
        meta_data = meta_data.append({'x':x,'y':y,'w':w,'h':h, 'Total_Faces': len(face),'Image_Name': i}, ignore_index=True)

```
